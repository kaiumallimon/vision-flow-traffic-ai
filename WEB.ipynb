```python
import cv2
import torch
import matplotlib.pyplot as plt
from ultralytics import YOLO

def get_explainable_features(image_path):
    """
    In 2026/2027, visualizing intermediate layers is the key to XAI.
    This function extracts the 'thought process' of the YOLO backbone.
    """
    model = YOLO('yolo11n.pt') # Use the .pt model for feature extraction
    
    # Run inference and get intermediate feature maps
    results = model.predict(image_path, visualize=True, save=True)
    
    # The 'visualize=True' argument automatically creates a 'runs/detect/predict/visualize' 
    # folder containing heatmaps of the model's layers!
    print(f"--- XAI Heatmaps saved to: {results[0].save_dir} ---")
    
    return results[0].save_dir

# Add this to your main flow
if __name__ == "__main__":
    test_img = "bus.jpg"
    xai_path = get_explainable_features(test_img)
    print(f"Check this folder for your Grad-CAM style explanations: {xai_path}")
```
